```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
hw = pd.read_csv('C:/Users/JYDIW/Documents/github/jydiw.github.io/assets/data/gender-height-weight.csv')
hw.columns = [
    'gender',
    'height-imp',
    'weight-imp',
    'height',
    'weight'
]
hw.shape
```

```python
np.random.seed(11)
sample = np.random.choice(9999, size=25)
hws = hw.iloc[sample]
```

```python
sns.set()
fig = sns.scatterplot(x='height', y='weight', data=hws, linewidth=0, legend=False)
plt.xlabel('height (cm)')
plt.ylabel('weight (kg)')
plt.title('weight vs. height')
plt.show(fig)
```

```python
np.mean(hws['weight'])
```

```python
fig = sns.scatterplot(x='height', y='weight', data=hws, linewidth=0, legend=False)

plt.plot((np.min(hws['height']), np.max(hws['height'])),
         (np.mean(hws['weight']), np.mean(hws['weight'])),
         color = 'darkred', lw=1
)

fig.text(152, 76.1, "y = 75.1")

plt.xlabel('height (cm)')
plt.ylabel('weight (kg)')
plt.title('weight vs. height')
plt.show(fig)
```

```python
lr = LinearRegression()

y = hws['weight']
X = hws[['height']]

lr.fit(X, y)
preds = lr.predict(X)
print(lr.coef_)
print(lr.intercept_)
```

```python
hws['preds'] = preds
```

```python
fig = sns.scatterplot(x='height', y='weight', data=hws, linewidth=0, legend=False)

plt.plot(hws['height'], hws['preds'],
         color = 'darkred', lw=1
)

fig.text(152, 76.1, 'y = 1.23x - 133')

plt.xlabel('height (cm)')
plt.ylabel('weight (kg)')
plt.title('weight vs. height')
plt.show(fig)
```

```python
fig = sns.scatterplot(x='height', y='weight', data=hws, linewidth=0, legend=False)

plt.plot((np.min(hws['height']), np.max(hws['height'])),
         (np.mean(hws['weight']), np.mean(hws['weight'])),
         color = 'darkred', lw=1
)

for _, row in hws.iterrows():
    plt.plot((row['height'], row['height']),
             (row['weight'], np.mean(hws['weight'])),
             color = 'blue',
             linestyle = '--',
             lw=1)

plt.xlabel('height (cm)')
plt.ylabel('weight (kg)')
plt.title('weight vs. height')
plt.show(fig)
```

```python
fig = sns.scatterplot(x='height', y='weight', data=hws, linewidth=0, legend=False)

plt.plot(hws['height'], preds,
         color = 'darkred', lw=1
)

for _, row in hws.iterrows():
    plt.plot((row['height'], row['height']),
             (row['weight'], row['preds']),
             color = 'blue',
             linestyle = '--',
             lw=1)

plt.xlabel('height (cm)')
plt.ylabel('weight (kg)')
plt.title('weight vs. height')
plt.show(fig)
```
